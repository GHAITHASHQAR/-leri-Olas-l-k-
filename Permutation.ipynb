Permütasyon Örneği : 3 öğrenci yuvarlak bir masa etrafında kaç farklı şekilde otururlar?

Solution : The permutation formula is used because the order is important to us.



```python
import math;
import itertools;


def InputKontrol(inputnumber):
    if not inputnumber.isdigit():
        return False
    else:
        return True


def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)

inputnumber=input("N students will be seated around a round table, please enter the number of students(N)!!")

kontrol_sonucu=InputKontrol(inputnumber)
if kontrol_sonucu==True:

    int_inputnumber=int(inputnumber)
    p_sonucu = factorial(int_inputnumber-1)
    print("%d student sits around the table in %d different ways(Permutation Result)" %(int_inputnumber,p_sonucu))

elif kontrol_sonucu==False:
    print("Enter a Number!!")
```

In how many ways can the same genders be arranged according to the number of boys and girls entered by the user?

##Explanation

Since boys can be lined up among themselves and girls among themselves, the factorial of the number of boys and girls is calculated and multiplied, respectively. Since boys and girls can change places as a group, the result is 2! multiplied by. So the formula is (Number of Boys)!.(Number of Girls)!.2! takes the form.



Erkekler kendi aralarında, kızlar kendi aralarında yer değiştirerek dizilebileceği için sırası ile erkek ve kız sayılarının faktoriyeli hesaplanarak çarpılır. Erkekler ve kızların da grup halinde yerleri değişebileceği için çıkan sonuç 2! ile çarpılır. Yani formül (Erkek Sayısı)!.(Kız Sayısı)!.2! şeklinde olur.

---



```python
def Calculatefactorial(i):    
    if i==1:       
        return 1    
    else: 
        return i * Calculatefactorial(i-1)
```

```python
def Calculatepermutation(j,k):
    
    l = Calculatefactorial(j)*Calculatefactorial(k)*Calculatefactorial(2)
    return l
```

```python
print("Enter the number of Boys and Girls to see how many different formations can be made, provided that Boys and Girls are side by side. ")   
 
num1 = int(input("Enter the Boys Number : "))
 
num2 = int(input("Enter the Girls Number: ")) 
 
print("\nSonuç:", Calculatepermutation(num1,num2)," can be arranged differently ")
```
