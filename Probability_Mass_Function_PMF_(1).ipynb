##**Probability mass function** :
 In probability and statistics, a probability mass function is a function that gives the probability that a discrete random variable is exactly equal to some valueSometimes it is also known as the discrete density function. The probability mass function is often the primary means of defining a discrete probability distribution, and such functions exist for either scalar or multivariate random variables whose domain is discrete.

A probability mass function differs from a probability density function (PDF) in that the latter is associated with continuous rather than discrete random variables. A PDF must be integrated over an interval to yield a probability.

##**Olasılık Kütle Fonksiyon (PMF)** :
Olasılık kuramı bilim dalında bir olasılık kütle fonksiyonu bir aralık rassal değişkenin olasılığının tıpatıp belli bir değere eşit olduğunu gösteren bir fonksiyondur. Olasılık kütle fonksiyonu, olasılık yoğunluk fonsiyonundan farklıdır, çünkü olasılık yoğunluk fonksiyonu yalnızca sürekli rassal değişkenler için tanımlanmış olup doğrudan doğruya olasılık değerini vermezler. Olasılık yoğunluk fonksiyonunun bir belli değer aralığı (yani a ve b değerleri aralığı) için integrali alinirsa bu rassal değişkenin belirlenen değer aralığı için olasılığını verir.

##**Example 1:** 
If a company wants to buy construction materials, different numbers of materials are taken as shown in the table below to purchase these materials; Let's show the distribution of the probability mass function for this example:download.png

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data ={'Construction_Materials':['wooden','Cement Bag','Glass Plate','Iron','Plastic'],
       'Material_Number':[300,450,250,530,120]}

df = pd.DataFrame(data)
df["possibility_pmf"] = df["Material_Number"]/df["Material_Number"].sum()
df

sns.barplot(df["Construction_Materials"],df["possibility_pmf"])
plt.show()
```

##**Example 2:** 
There are 9 male and 6 female in a Hospital. Two doctors on duty will be chosen at random. Find the discrete probability distribution of the number of female doctors.


M: Male F: Female

**1. Durum (D1): Hiç bayan doktor gelmeme durumu**

 **MM**  P(X=D1) = (9/15) * (8/14) = 0.342

<hr>


**2. Durum (D2): Bir bayan doktor gelme durumu**

**MF - KE** P(X=D2) =  ((9/15) * (6/14)) + ((6/15) * (9/14)) =  0,514

<hr>


**3. Durum (D3): İki bayan doktor gelme durumu**

**FF** P(X=D3) = (6/15) * (5/14) = 0,143

```python
import numpy as np


MALE_Num=9
FEMALE_Num=6
Total_Num=15

#calculation the PMF
pmf = []

D1= (MALE_Num/Total_Num) * ((MALE_Num-1)/(Total_Num-1)) 
pmf.append(D1)

D2=((MALE_Num/Total_Num) * (FEMALE_Num/(Total_Num-1))) + ((FEMALE_Num/Total_Num) * (MALE_Num/(Total_Num-1)))
pmf.append(D2)

D3= (FEMALE_Num/Total_Num) *  ((FEMALE_Num-1)/(Total_Num-1))
pmf.append(D3)

x=['D1', 'D2','D3']



np.array(pmf).sum()





```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

plt.bar(x, pmf);
plt.title("Olsılık Kütle Fonksiyonu");
```
